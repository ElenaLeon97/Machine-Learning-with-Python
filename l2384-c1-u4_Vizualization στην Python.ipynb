```python
# 1.0 Εισαγωγή
## 1.1 Γραφική απεικόνιση (plot) συνημιτόνου
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib inline # Η επιλογή %matplotlib inline στον κώδικα διασφαλίζει ότι εμφανίζονται όλα τα plots μας καθώς εκτελούμε κάθε κελί. 
```

```python
# creating plot values
x = np.linspace (0,10) #x-axis/time scale
y = np.cos(x)
```

```python
plt.plot(x,y)
```

```python
# 1.2 Πολλές συναρτήσεις σε ένα γράφημα
z = np.sin(x)
plt.plot(x,y,x,z)
```

```python
# 1.3 Ετικέτες και τίτλος γραφήματος
plt.plot(x,y,x,z)
plt.xlabel('time')
plt.ylabel('magnitude')
plt.title('sine and cosine waves')
```

```python
# 1.4 Η μέθοδος legend()
plt.plot(x,y,x,z)
plt.xlabel('time')
plt.ylabel('magnitude')
plt.title('sine and cosine waves')
plt.legend(['cos','sin'],loc=0)
```

```python
# 1.5 Παράλληλα Plots
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('cosine')
plt.subplot(1,2,2)
plt.plot(x,z)
plt.title('sine')
plt.tight_layout() # διασφαλίζει ότι όλα τα subplots βρίσκονται σε επαρκείς αποστάσεις μεταξύ τους.
```

```python
# 1.6 Χρώματα της γραμμής στα plots και markers
plt.subplot(1,2,1)
plt.plot(x,y, 'r-x')
plt.title('cosine')
plt.subplot(1,2,2)
plt.plot(x,z, 'b-o')
plt.title('sine')
plt.tight_layout()
# 1.7 Αποθήκευση και αντιγραφή των plots
plt.savefig('figure.jpg' , DPI=300)
```

```python
# 2. Seaborn
## 2.1 Μεταφόρτωση ενός συνόλου δεδομένων της seaborn και σχεδιασμός τάσεων (trends)
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
## 2.3 Οπτικοποίηση τάσεων (trends).
### 2.3.1 Γραφική παράσταση κατανομής (distribution plot)
sns.displot(tips['total_bill'], bins = 30, kde=False)
plt.show()
```

```python
### 2.3.2 Σχέση μεταξύ δύο μεταβλητών
sns.relplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.show()
```

```python
### 2.3.3 Η μέθοδος pairplot()
sns.pairplot(tips, hue='sex', palette = 'coolwarm')
plt.show()
```

```python
### 2.3.4 Παράσταση ιστογράμματος (histogram plot) με τη μέθοδο hist()
tips.hist()
plt.show()
```

```python
### 2.3.5 Η μέθοδος countplot()
sns.countplot(x='sex', data=tips, hue='sex')
plt.show()
```

```python
# 3. Οπτικοποιηση Δεδομενων απο Data Frame της Pandas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## 3.1 Δημιουργία Data Frame δεδομένων
dico = {'A':np.random.rand(5), 'B':np.random.rand(5), 'C':np.random.rand(5), 'D':np.random.rand(5)}
df = pd.DataFrame(dico)
df
```

```python
## 3.2 Δημιουργία bar plot και μορφή στοίβας (stacked)
df.plot.bar() #ο άξονας x έχει τον δείκτη γραμμής και ο άξονας y δείχνει την τιμή σε κάθε στήλη ανά δείκτη.
plt.show()
```

```python
df.plot.bar(stacked = True, alpha = 0.8)
plt.show()
```

```python
## 3.3 Γραμμική γραφική παράσταση (line plot)
df.plot.line(y=['B','C'])
plt.show()
```
