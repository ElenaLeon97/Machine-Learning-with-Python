### Bubble sort

```python
L = [13,40,50,50,90,18,30,50,30,70]
```

```python
N=len(L) # πλ΄΄ηθος στοιχείων λίστας
```

```python
print('Αρχική λίστα: ', L)
```

```python
for i in range (N-1):
    for j in range (N-1-i):
        if L[j+1] < L[j]:
            L[j],L[j+1]=L[j+1],L[j]
```

```python
print('Ταξινομημένη λίστα: ', L)
```

```python
v=stat.pvariance(L)
```

```python
v
```

```python
stdev=stat.pstdev(L)
```

```python
stdev
```
